```python
words = open("names.txt", "r").read().splitlines()
```

```python
import numpy as np
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
%matplotlib inline
import itertools
```

```python
chars = sorted(list(set("".join(words))))
chars += "."
char_comb = [p for p in itertools.product(chars, repeat=2)]
stoi = {s:i for i, s in enumerate(char_comb)}
itos = {i:s for s, i in stoi.items()}
ytoi = {s:i+1 for i,s in enumerate(chars[:-1])}
ytoi["."] = 0
itoy = {i:s for s,i in ytoi.items()}
```

```python
xs, ys = [], []
for w in words:
    chs = ["."] + list(w) + ["."]
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        xs.append(stoi[(ch1, ch2)])
        ys.append(ytoi[ch3])


xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
```

```python
W = torch.randn((729, 27), requires_grad=True)
```

```python
itos
```

```python
for k in range(10):
    #forward pass
    xenc = F.one_hot(xs, num_classes=729).float()
    logtis = xenc @ W
    count = logtis.exp()
    probs = count / count.sum(1, keepdim=True)
    #calc loss
    loss = -probs[torch.arange(num), ys].log().mean()
    if not k%1:
        print(k, loss)
    W.grad = None
    loss.backward()
    W.data -= 200 * W.grad
```
